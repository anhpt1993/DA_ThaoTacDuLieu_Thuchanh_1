# 0. Import các thư viện cần thiết

```python
import pandas as pd
import numpy as np
```

# 1. Đọc file CSV

```python
df = pd.read_csv("https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2022/01/20125029/shopeep_koreantop_clothing_shop_data.csv", encoding = "ISO-8859-1")
df.head()
```

```python
# Lọc các cột dữ liệu "join_month", "join_day", "join_year", "shop_location", "rating_bad"
# "rating_good" & "rating_normal"

df = df.loc[:, ["join_month", "join_day", "join_year", "shop_location", "rating_bad", "rating_good", "rating_normal"]]
df.head(10)
```

# 2. Tạo cột 'rating' tính điểm cho mỗi cửa hàng dựa vào thông tin các cột rating_bad, rating_good, rating_normal 
- theo công thức sau: rating = rating_good x 2 + rating_normal - rating_bad x 3

```python
df["rating"] = df["rating_good"] * 2 + df["rating_normal"] - df["rating_bad"] * 3
df.head(10)
```

```python
df.info()
```

# 3. Ghép 3 cột join_month, join_day, join_year thành cột mới có tên 'date' nhận giá trị có dạng: "join_month join_day,join_year"

```python
df["date"] = df["join_month"] + " " + df["join_day"].apply(str) + "," + df["join_year"].apply(str)
df.head(10)
```

# 4. Thêm cột mới theo điều kiện

```python
df.info()
```

```python
# Thêm cột new có giá trị True nếu join_year = 2021 và False trong trường hợp còn lại
df["new"] = df["join_year"] == 2021
df.head(10)
```

```python
# Nếu có 2 lựa chọn ta sử dụng np.where
# Thêm cột rate có giá trị good nếu rating_good >= 50000,  bad trong trường hợp còn lại
df["rate"] = np.where(df["rating_good"] >= 50000, "good", "bad")
df.head(10)
```

```python
# Nếu có nhiều hơn 2 lựa chọn ta sử dụng np.select
# Thêm cột flag tặng cờ cho các cửa hàng, flag nhận các giá trị như sau:
# blue khi rating_good >= 30000 và rating_bad <= 100
# yellow khi 10000 <= rating_good < 30000 và 100 < rating_bad <= 1000
# red khi rating_good < 10000
# black đối với các trường hợp còn lại

conditions = [(df.rating_good >= 30000) & (df.rating_bad <= 100),
              (10000 <= df.rating_good) & (df.rating_good <= 30000) & (100 < df.rating_bad) & (df.rating_bad <= 1000),
              (df.rating_good < 10000)]
choices = ["blue", "yellow", "red"]
df["flag"] = np.select(conditions, choices, default = "black")
df.head(10)
```
